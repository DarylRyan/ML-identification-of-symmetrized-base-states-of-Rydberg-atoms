```python
import matplotlib.pyplot as plt

x = [3,7,10]

lin4 = [1.00, 1.00, 0.97 ] 
lin4err = [0.01, 0.01, 0.02]
C4 = [1.00, 1.00, 0.99]
C4err = [0.01, 0.01,0.02]
linC4 = [0.98, 0.99, 0.88]
linC4err = [0.01, 0.01, 0.03]

rfclin4 = [1.00, 1.00, 0.97] 
rfclin4err = [0.01, 0.01, 0.02]
rfcC4 = [1.00, 1.00, 0.99]
rfcC4err = [0.01, 0.01, 0.02 ]
rfclinC4 = [0.98, 0.98, 0.93]
rfclinC4err =[0.01, 0.02, 0.02]

# plt.plot(x,lin4,)

# plt.plot(x,C4,)

plt.errorbar(x,lin4,
            yerr=lin4err,
             
            ms=4, 
            capsize=5,
            capthick=2,
            label='linear')


plt.errorbar(x,C4,
            yerr=C4err,
 
            ms=4, 
            capsize=5,
            capthick=2,
            label='C4')

plt.errorbar(x,linC4,
            yerr=linC4err,
             
            ms=4, 
            capsize=5,
            capthick=2,
            label='linear + C4')

# # Set x limits
plt.xlim(0, x[-1]*1.05)
# Set y limits
plt.ylim(0.84, 1.019)

# plt.title('Plot of accuracy against N')
plt.xlabel('Noise (%)')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.hlines(1,0, x[-1]*1.05, colors='r', linestyles='dashed')
# plt.savefig(file_path_to_save_plot, dpi=300, bbox_inches='tight')
plt.title('SVM')
plt.show()
```
