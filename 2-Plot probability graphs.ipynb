# Plot all data files in a folder

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Import data

```python
# Folder that contains the data to be plotted
folder_of_data_to_plot = r"C:\Users\dell\0 Jupyter Notebooks - python\Dr Heejeong - UM Research\Project 1 - Machine learning of probabconfiguration from probability graphs\12.2.2021 KAIST Data\data 5atoms lin N300 V13 update"
```

# Determine if plots will be saved

```python
save_plots = False # set to True or False
saved_file_type = '.pdf' # select file type of saved plots ('.png' , '.pdf', etc.)
folder_of_saved_plots = r"C:\Users\dell\0 Jupyter Notebooks - python\Dr Heejeong - UM Research\18.2.2021 Continuous graphs of probability\N1 all atom data with NNN T201\1-6 atoms lin chains"
```

### List of files that will be plotted

```python
data_files = os.listdir(folder_of_data_to_plot)
data_files = sorted(data_files)
data_files
```

```python
for i in data_files:
    index = data_files.index(i)
    #store data in df
    df = pd.read_csv(folder_of_data_to_plot + '\\' + i, header=None)

    #accessing values for x and y axis
    x = df.iloc[:,0]
    y = df.iloc[:,1]

    plt.figure(index)
    plt.title(i[1:2])
    plt.xlabel('Time ($\mu$s)')
    plt.ylabel('Probability')
#     plt.xlim(0, 1.03)
#     plt.ylim(0.0, 1.05) #set range of y axis
    plt.grid(True)
    plt.plot(x,y)
    if save_plots:
        plt.savefig(folder_of_saved_plots + i[:-4] + saved_file_type, dpi=300, bbox_inches='tight')
```

```python
number_of_rows = 2
number_of_columns = 2

fig, axs = plt.subplots(ncols = number_of_columns, nrows = number_of_rows, sharey=True, sharex=True)
# fig, axs = plt.subplots(ncols = 2, nrows = 3, sharey=True, sharex=True)

axs = axs.ravel()

for i in data_files:
    index = data_files.index(i)
    #store data in df
    df = pd.read_csv(folder_of_data_to_plot + '\\' + i, header=None)
    #accessing values for x and y axis
    x = df.iloc[:,0]
    y = df.iloc[:,1]
#     axs[index].set_title(i[1:2])
    axs[index].plot(x,y, label=i[1:2], linewidth=1)
#     axs[index].legend(loc=3)
    axs[index].text(x=0.5,y =0.9, s=i[1:2])
#     fontsize or size	float or {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
    axs[index].grid(True)
    
    
    if save_plots:
        plt.savefig(folder_of_saved_plots + i[:-4] + saved_file_type, dpi=300, bbox_inches='tight')

axs[0].set_ylabel('Probability')
axs[2].set_ylabel('Probability')
# axs[3].set_ylabel('Probability')
# axs[4].set_ylabel('Probability')
# axs[5].set_ylabel('Probability')
# axs[6].set_ylabel('Probability')
# axs[8].set_ylabel('Probability')

# axs[0].set_xlabel('Time ($\mu$s)')
# axs[1].set_xlabel('Time ($\mu$s)')

axs[2].set_xlabel('Time ($\mu$s)')
axs[3].set_xlabel('Time ($\mu$s)')

# axs[4].set_xlabel('Time ($\mu$s)')
# axs[5].set_xlabel('Time ($\mu$s)')

# axs[6].set_xlabel('Time ($\mu$s)')
# axs[7].set_xlabel('Time ($\mu$s)')
# axs[8].set_xlabel('Time ($\mu$s)')
        
# axs[8].set_xlabel('Time ($\mu$s)')
# axs[9].set_xlabel('Time ($\mu$s)')
# axs[10].set_xlabel('Time ($\mu$s)')
# axs[11].set_xlabel('Time ($\mu$s)')


def cm_to_inch(value):
    return value/2.54

# a single axes should be 15 cm wide by 10 cm high
fig.set_figheight(cm_to_inch(10*number_of_rows))
fig.set_figwidth(cm_to_inch(15*number_of_columns))
plt.subplots_adjust(wspace=0, hspace=0)

# fig.suptitle(t='(d)', x=0.1, y=0.93,fontsize=16)

plt.show()
```
